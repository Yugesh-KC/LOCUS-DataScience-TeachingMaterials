```python
x = np.linspace(0,1000,200)  #creates equally spaced 100 integers between 0 and 1000
x_radians = x * (np.pi / 180)
y_sin = np.sin(x_radians)
plt.plot(x,y_sin)
plt.show()
```

```python
y_cos = np.cos(x_radians)
plt.plot(x,y_cos)
plt.show()
```

```python
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.legend(['sin','cos'])  #first line plotted gets first value of legend and so on
plt.title('Sin and Cosine')
plt.show()
```

```python
df = pd.read_csv('mpg.csv')
```

```python

df.head()
```

```python
df = df[df['horsepower'] != '?']
df['horsepower']=df['horsepower'].astype('float')
```

```python
plt.plot(df['weight'],df['mpg'])
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.title('MPG vs Weight')
plt.show()
```

Line plot plots each datapoints and connects them with a line in the order they appear. If a value in x-axis decreases the point is towards the left line connects the plot from left right to left which results in figures like this

```python
df1 = df.sort_values('weight')
```

```python
plt.plot(df1['weight'],df1['mpg'])
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.title('MPG vs Weight')
plt.show()
```

The figure looks better now. It tells us that the mileage of a vehicle decreases as weight increase. But it still is not the best way to visualise these columns. A line plot assumes continuous progression, (like time series), which is not the case here.

```python
plt.scatter(df['weight'],df['mpg'])
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.title('MPG vs Weight')
plt.show()
```

```python
plt.scatter(df['weight'],df['mpg'],s = df['horsepower']*0.5)
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.title('MPG vs Weight- Sized by Horsepower ')
plt.show()
```

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.scatter([5,6,7,8],[9,10,11,12])
plt.legend(['First','Second'])
plt.show()
```

```python
unique_cylinders = df['cylinders'].unique()
for i in unique_cylinders:
    df1 = df[df['cylinders'] == i]
    plt.scatter(df1['weight'],df1['mpg'],label = f'cylinders = {i}') #Instead of providing a list in plt.legend(), we can simply write the label in each plot

plt.xlabel('Weight')
plt.ylabel('MPG')
plt.title('MPG vs Weight- With cylinders ')
plt.legend()
plt.show()
```

```python
model_year_freq= df['model_year'].value_counts()
model_years = model_year_freq.index
print(model_year_freq)
print(model_years)
```

```python
plt.hist(df['mpg'], bins=5)   #bins divide the data equally
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('Histogram of MPG')
plt.show()
```

The figure is still difficult to interpret. Let's add ticks and edge colors

```python
counts, bins, patches = plt.hist(df['mpg'], bins=5,edgecolor='black')
plt.xticks(bins)
# plt.yticks(counts)   #uncomment and see what happens
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('Histogram of MPG')
plt.show()
```

```python
counts, bins, patches = plt.hist(df['mpg'], bins=5,edgecolor='black',cumulative = True)
plt.xticks(bins)
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('Histogram of MPG')
plt.show()
```

```python
box=plt.boxplot([df['mpg']])
plt.xticks([1],['MPG'])
plt.show()
```

```python
box = plt.boxplot(df['mpg'], showmeans=True)
print(box)


median = box['medians'][0].get_ydata()[0]
print(f"Median: {median}")

box_coords = box['boxes'][0].get_ydata()
q1 = min(box_coords)
q3 = max(box_coords)
print(f"Q1: {q1}")
print(f"Q3: {q3}")

whisker_low = box['whiskers'][0].get_ydata()[1]
whisker_high = box['whiskers'][1].get_ydata()[1]
print(f"Lower whisker: {whisker_low}")
print(f"Upper whisker: {whisker_high}")

cap_low = box['caps'][0].get_ydata()[0]
cap_high = box['caps'][1].get_ydata()[0]
print(f"Cap low: {cap_low}")
print(f"Cap high: {cap_high}")


mean = None
if 'means' in box and box['means']:
    mean = box['means'][0].get_ydata()[0]
    print(f"Mean: {mean}")

yticks = sorted(set([median, q1, q3, whisker_low, whisker_high, cap_low, cap_high] + ([mean] if mean else [])))
plt.yticks(yticks)

plt.xticks([1],['MPG'])

plt.show()

```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10), dpi=100)

axes[0, 0].bar(model_years,model_year_freq)
axes[0, 0].set_title('Bar Graph: Model Years')

axes[0, 1].scatter(df['weight'], df['mpg'])
axes[0, 1].set_title('Scatter Plot: Weight vs MPG')

axes[1, 0].hist(df['mpg'], bins=5, edgecolor='black')
axes[1, 0].set_title('Histogram of MPG')

axes[1, 1].boxplot(df['mpg'])
axes[1, 1].set_title('Boxplot of MPG')

fig.suptitle('Subplots Example', fontsize=16)


plt.show()
```

```python

fig = plt.figure(figsize=(10, 10), dpi=100)

# Create axes manually with precise position and size
ax1 = fig.add_axes([0.1, 0.55, 0.35, 0.35])  # left, bottom, width, height (in terms of fraction of figure width/height)
ax2 = fig.add_axes([0.55, 0.55, 0.35, 0.35])
ax3 = fig.add_axes([0.1, 0.1, 0.35, 0.35])
ax4 = fig.add_axes([0.55, 0.1, 0.35, 0.35])

ax1.bar(model_years, model_year_freq)
ax1.set_title('Bar Graph: Model Years')

ax2.scatter(df['weight'], df['mpg'])
ax2.set_title('Scatter Plot: Weight vs MPG')

ax3.hist(df['mpg'], bins=5, edgecolor='black')
ax3.set_title('Histogram of MPG')

ax4.boxplot(df['mpg'])
ax4.set_title('Boxplot of MPG')

fig.suptitle('Manual multi-plot figure', fontsize=16)

plt.show()
```
